```python
'''
引入套件和資料集
官方文件：https://seaborn.pydata.org/generated/seaborn.lineplot.html
注意：seaborn 要較新的版本才有 lineplot
'''
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.__version__)

flights = sns.load_dataset("flights")
flights.head()
```

```python
# seaborn 的優點就是可以直接取用 dataframe 的欄位做畫圖的參數
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
# ci：設定信賴區間，不要顯示則填 None
# err_style：bars, band
sns.lineplot(data=flights, x="year", y="passengers", ci=95, err_style='bars')
```

```python
'''
legend="full"：hue 秀出所有類別
使用 matplotlib 將 xticks 做旋轉
色盤選項：https://seaborn.pydata.org/tutorial/color_palettes.html
'''

sns.lineplot(data=flights, x="month", y="passengers", hue='year', legend="full", palette='Paired')
plt.xticks(rotation=90)
```

```python
# 直接計算資料點數，無須再使用 groupby 等函數
sns.countplot(data=flights, x="year")
```

```python
'''
col：將欄位展開成多個圖
col_wrap：客製行數量
並將多個 subplot xticks 做旋轉
'''

fig = sns.relplot(data=flights, x="month", y="passengers", col='year', col_wrap=4, kind='scatter')
for ax in fig.axes:
    ax.tick_params(labelrotation=90)
```

```python
# 只要將參數放的欄位做更動就可以將圖形轉向

flights_1960 = flights.query("year == 1960")
sns.barplot(data=flights_1960, y='month', x='passengers')
```
