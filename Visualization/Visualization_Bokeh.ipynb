```python
from bokeh.plotting import figure, show, output_file # 存成 html
from bokeh.io import output_notebook # 直接 show 在 jupyter notebook console
```

```python
'''
line plot
output_file + show() 在 py 檔同層存成 line.html
需開啟檔案才能看到圖型
'''
output_notebook()
x1 = list(range(10))
y1 = list(range(20,0,-2))
x2 = list(range(-3,5))
y2 = list(range(8))


output_file('line.html') #  
p = figure(width=800, height=400, title='line plot')
p.line(x1, y1, line_width=5, line_color='red', line_alpha=0.5, line_dash=[12,3], legend='red')
p.line(x2, y2, line_width=8, line_color='green', line_alpha=0.3, line_dash=[18,1], legend='green')
p.title.text_color='blue'
p.title.text_font_size='16pt'
p.xaxis.axis_label='time'
p.yaxis.axis_label='value'
p.xaxis.axis_label_text_color='brown'
p.yaxis.axis_label_text_color='brown'
show(p)
```

```python
'''
scatter plot with circle and square
並不需要另外設置中文字的問題
'''
output_notebook()
p = figure(width=800, height=400, title="統計圖")
p.title.text_font_size = "18pt"
p.xaxis.axis_label = "X 軸"
p.yaxis.axis_label = "y 軸"
listx = [1,5,7,9,13,16]
listy = [15,50,80,40,70,50]
sizes=[10,20,30,30,20,10]
colors=["red","blue","green","pink","violet","gray"]

#sizes=25  #所有點相同大小
#colors="red"  #所有點相同顏色
p.circle(listx, listy, size=sizes, color=colors, alpha=0.5)
p.square(listx, listy, size=sizes, color=colors, alpha=0.5)
show(p)
```

```python
'''
pie chart with explode by calculating
explode : one wedge by offsetting its center
'''
import math
from math import pi
from bokeh.plotting import figure, show

output_notebook()

p = figure(x_range=(-1.2, 1.2), y_range=(-1.2, 1.2))

p.wedge(x=0 , y=0, radius=1, start_angle=0, end_angle=0.75*pi, color="#9A8FBF")
p.wedge(x=0 , y=0, radius=1, start_angle=0.75*pi, end_angle=1.05*pi, color="#DCE4F2")
p.wedge(x=0 , y=0, radius=1, start_angle=1.05*pi, end_angle=1.95*pi, color="#B0CDD9")

cal = (1.95*pi+2*pi)/2
adj_x = 0.15*math.cos(cal)
adj_y = -0.15*math.sin(cal)
p.wedge(x=adj_x , y=-adj_y, radius=1, start_angle=1.95*pi, end_angle=2*pi, color="#8C6056")

show(p) 
```
